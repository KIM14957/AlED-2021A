```python
##1label encoding : 순서가 있을 때인데 school은 ??? --> one-hot 인코딩 필요

import numpy as np
from sklearn.preprocessing import LabelEncoder,OneHotEncoder


df['school']

## This function converts GPand MS to 0&1
def convert(school):
    if(school=='GP'):
        return 1
    else :
        return 0

df['school'] = df['school'].apply(convert)

df['school']

```

```python
##1label encoding : 순서가 있을 때인데 school은 ??? --> one-hot 인코딩 필요

from sklearn import preprocessing

label_encoder=preprocessing.LabelEncoder()
df['school']=label_encoder.fit_transform(df['school'])
df['school']
```

```python
##2 one-hot encoding

## get_dummies()  사용한 원핫 ##

from sklearn.preprocessing import OneHotEncoder 

y=pd.get_dummies(df.school,prefix='school')

y.head()

```

```python
# y을 원래 df에 add  : df --> dfOneHot

# dfOneHot=pd.DataFrame(y, columms=["school_"+str(int(i)) for in in range(data.shape[1])])

df=pd.concat([df, y],axis=1)

df=df.drop(['school'], axis=1)
df.head()

```
